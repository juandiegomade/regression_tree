```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.tree import DecisionTreeRegressor
from sklearn import tree
```

```python
df_raw = pd.read_csv('../data/raw/polls.csv')
df_raw.info()
```

```python
df_processed = df_raw[['day','margin'].copy()]
df_processed.sample(10)
```

```python
df = df_processed.copy()
df.to_csv('../data/final/polls_final.csv')
```

```python
fig,ax = plt.subplots()
ax.plot(df['day'],df['margin'])
plt.show()
```

```python
model = DecisionTreeRegressor(min_samples_split=20,ccp_alpha=0)
model.fit(df[['day']],df['margin'])
```

```python
y_hat = model.predict(df[['day']])
```

```python
fig,ax = plt.subplots()
ax.plot(df['day'],y_hat)
plt.show()
```

```python
fig = plt.figure(figsize=(15,15))
tree.plot_tree(model,feature_names=['day'])
plt.show()
```
